#### Q1: What is the distribution of ages at enrollment of students in the dataset?

```python
# create histogram of 'Age at enrollment' column
plt.hist(df['Age at enrollment'], bins=20)
plt.xlabel('Age at enrollment')
plt.ylabel('Count')
plt.title('Distribution of Ages at Enrollment')
plt.show()
```

#### Q2: How does the distribution of the number of approved curricular units in the 1st semester differ between male and female students?

```python
# create box plot of 'Curricular units 1st sem (approved)' column, grouped by 'Gender
sns.boxplot(x='Gender', y='Curricular units 1st sem (approved)', data=df)
plt.xlabel('Gender')
plt.ylabel('Curricular units 1st sem (approved)')
plt.title('Distribution of Approved Curricular Units in the 1st Semester by Gender')
plt.show()
```

#### Q3:How is the relationship between the number of credited curricular units in the 1st semester and the grade received in that semester?

```python
# create scatter plot of 'Curricular units 1st sem (credited)' vs. 'Curricular units 1st sem (grade)'
plt.scatter(df['Curricular units 1st sem (credited)'], df['Curricular units 1st sem (grade)'])
plt.xlabel('Curricular units 1st sem (credited)')
plt.ylabel('Curricular units 1st sem (grade)')
plt.title('Relationship between Credited Curricular Units in the 1st Semester and Grade Received')
plt.show()
```

#### Q4: Is there a correlation between the number of credited curricular units in the 1st semester and the number of evaluations in that semester?

```python
# create scatter plot of 'Curricular units 1st sem (credited)' vs. 'Curricular units 1st sem (evaluations)'
plt.scatter(df['Curricular units 1st sem (credited)'], df['Curricular units 1st sem (evaluations)'])
plt.xlabel('Curricular units 1st sem (credited)')
plt.ylabel('Curricular units 1st sem (evaluations)')
plt.title('Relationship between Credited Curricular Units in the 1st Semester and Evaluations')
plt.show()
```

#### Q5: How does previous qualification affect the number of credited curricular units in the first semester?

```python
#We can compare the mean number of credited curricular units in the first semester for each level of previous qualification.
# grouping data by previous qualification and calculating the mean of credited curricular units in the first semester
mean_units_1st_sem = df.groupby("Previous qualification")["Curricular units 1st sem (credited)"].mean()

# plotting the mean number of credited curricular units in the first semester for each level of previous qualification
plt.bar(mean_units_1st_sem.index, mean_units_1st_sem.values)
plt.xlabel("Previous Qualification")
plt.ylabel("Mean Curricular units 1st sem (credited)")
plt.title("Effect of Previous Qualification on Curricular units 1st sem (credited)")
plt.show()
```

#### Q6:How does scholarship holder status affect the number of approved curricular units in the first and second semesters?

```python
# Group by scholarship holder status and get the mean number of approved curricular units in each semester
scholarship_approvals_df = df.groupby("Scholarship holder")[["Curricular units 1st sem (approved)", "Curricular units 2nd sem (approved)"]].mean()

# Plot the results
scholarship_approvals_df.plot(kind="bar", figsize=(10, 6))
plt.xlabel("Scholarship holder")
plt.ylabel("Mean number of approved curricular units")
plt.title("Mean number of approved curricular units by scholarship holder status in each semester")
plt.show()
```

## Inferences and Conclusion


**Summary:**

The data analysis was conducted on the 'Higher Education Predictors of Student Retention' dataset, which includes various student attributes such as marital status, application mode, previous qualification, mother and father's occupation, tuition fees, gender, age, etc. The target column in the dataset represents whether a student was retained or not.


**Insights:**

1. The distribution of ages at enrollment of students in the dataset is positively skewed, with the majority of students being between 20-25 years old.

2. The distribution of the number of approved curricular units in the 1st semester showed a slightly higher mean for female students compared to male students, but overall, the distributions were similar.

3. There is a positive relationship between the number of credited curricular units in the 1st semester and the grade received in that semester, indicating that students who take more classes tend to perform better.

4. There is a moderate positive correlation between the number of credited curricular units in the 1st semester and the number of evaluations in that semester, suggesting that students who take more classes also receive more evaluations.

5. The previous qualification of the students has a moderate positive impact on the number of credited curricular units in the first semester. Students with higher previous qualifications tend to take more classes.

6. Scholarship holder status has a moderate positive impact on the number of approved curricular units in both the first and second semesters. Scholarship holders tend to take more classes and perform better.
